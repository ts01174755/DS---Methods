```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame({"X": ['Wendy', 'Wendy', 'Fred','Fred'], "Y": ['a', 'b', 'c','e']})
df1
```

```python
ft_dummy = pd.get_dummies(df1['Y'],prefix='Dummy',prefix_sep='_')
ft_dummy
```

```python
df_ftdummy = df1[['X']].join(ft_dummy)
df_ftdummy
```

```python
df_ftdummy_unit = df_ftdummy.groupby(by=['X']).max()
df_ftdummy_unit
```

```python
df1 = df1.merge(df_ftdummy_unit,how='left',left_on='X',right_on='X')
df1
```
